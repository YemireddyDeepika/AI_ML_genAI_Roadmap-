**LinearRegression and PolynomialRegression**

1.Importing Libraries

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
```

2.Data Set

```python
data = { "Experience" : [1,2,3,4,5,6],
          "Salary" : [20000,30000,45000,60000,65000,70000]} 
df = pd.DataFrame(data)
x = df[["Experience"]]
y = df["Salary"]
```

3.LinearRegression

```python
lin_model = LinearRegression()
lin_model.fit(x,y)
y_pred_lin = lin_model.predict(x)

print("Linear Regression Results:\n")
print("MSE of LinearRegression:",mean_squared_error(y,y_pred_lin))
print("R2 Score of LinearRegression:",r2_score(y,y_pred_lin))
```

4.PolynomialRegression

```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

poly_model = LinearRegression()
poly_model.fit(x_poly,y)

y_pred_poly = poly_model.predict(x_poly)

print("Ploynomail reggression:\n")
print("MSE of Polynomial Regression:\n",mean_squared_error(y,y_pred_poly))
print("R2 of Polynomail regression:\n",r2_score(y,y_pred_poly))
```

5.Prediction for New Value

```python
new_exp = pd.DataFrame({"Experience":[7]})
print("\nPrediction for Experience = 7")
print("Linear Regression Prediction:",lin_model.predict(new_exp))
print("Polynomial Regression Preduction:",poly_model.predict(poly.transform(new_exp)))
```

6.Visual Comparison

```python
x_sorted = x.sort_values(by = "Experience")
plt.scatter(x,y,color="black",label="Actual Data")
plt.plot(x_sorted,lin_model.predict(x_sorted),color="red",label="Linear Regression")
plt.plot(x_sorted,poly_model.predict(poly.transform(x_sorted)),color="blue",label="Polynomial Rergression")
plt.xlabel("Experience")
plt.ylabel("Marks")
plt.title("Linear Vs Polynomial Regression")
plt.legend()
plt.show()
```
