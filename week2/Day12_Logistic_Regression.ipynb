**Mini Task**

Predict whether a student Pass(1) or Fail(0) based on Study Hours

1.Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score,confusion_matrix
import matplotlib.pyplot as plt
```

2.Dataset

```python
data ={"Study_Hours" : [1,2,3,4,5,6,7,8],
       "Pass" : [0,0,0,0,1,1,1,1]}
df = pd.DataFrame(data)
```

3.Model

```python
x = df[["Study_Hours"]]
y = df["Pass"]
log_model = LogisticRegression()
log_model.fit(x,y)
```

4.Prediction

```python
y_pred = log_model.predict(x)
print("Prediction:",y_pred)
```

5.Evaluation

```python
accuracy = accuracy_score(y,y_pred)
cm = confusion_matrix(y,y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:",cm)
```

6.Prediction for New Student

```python
new_hours = pd.DataFrame({"Study_Hours":[4.5]})
prediction = log_model.predict(new_hours)
probability = log_model.predict_proba(new_hours)
if probability[0][1]>= 0.45:
    
#print("Prediction (0 = Fail, 1 = Pass):",prediction[0])
#print("Probability of Fail:",probability[0][0])
#print("Probability of Pass:",probability[0][1])
    print("Final Decision:Pass")
else:
    print("Final Decision:Fail")
```

7.Visualization

```python
plt.scatter(x,y,color="black",label = "Actual Data")
plt.plot(x,log_model.predict_proba(x)[:,1],color = "blue",label = "Sigmoid Curve")
plt.xlabel("Sudy_Hours")
plt.ylabel("Pass")
plt.legend()
plt.show()
```
