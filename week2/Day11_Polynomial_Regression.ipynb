**Polynomial Regression**

1.Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

2.Dataset

```python
data = {"Hours_Studied" : [1,2,3,4,5],
        "Marks" : [40,50,65,85,90]}
df = pd.DataFrame(data)
x=df[["Hours_Studied"]]
y=df["Marks"]
```

3.Create Polynomial Feature

```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
```

4.Create Model

```python
model = LinearRegression()
model.fit(x_poly,y)
```

5.Prediction

```python
new_exp = pd.DataFrame({"Hours_Studied":[6]})
prediction = model.predict(poly.transform(new_exp))
print("Predicted Marks:", prediction)
```

```python
y_pred = model.predict(x_poly)
r2 = r2_score(y,y_pred)
print("R2 Score:",r2)
```

**Mini Task**

```python
#Create Dataset
data1 = { "Experience" : [1,2,3,4,5,6],
          "Salary" : [20000,30000,45000,60000,65000,70000]} 
df1=pd.DataFrame(data1)
X = df1[["Experience"]] 
Y = df1["Salary"]

#Linear Regression
lin_model = LinearRegression()
lin_model.fit(X,Y)
lin_new_exp = pd.DataFrame({"Experience":[7]})
lin_pred = lin_model.predict(lin_new_exp)
print("Linear Regression Prediction:\n",lin_pred)
Y_pred_train = lin_model.predict(X)
lin_mse = mean_squared_error(Y,Y_pred_train)
lin_r2 = r2_score(Y,Y_pred_train)
print("MSE of Linear Regression:\n",lin_mse)
print("R2 of Linear Regression:\n",lin_r2)
```

```python
#Polynomial Regression
poly_model = PolynomialFeatures(degree=2)
X_polyn = poly_model.fit_transform(X)
lin_model.fit(X_polyn,Y)
Y_pred = lin_model.predict(X_polyn)
poly_new_exp = pd.DataFrame({"Experience" : [7]})
poly_pred = lin_model.predict(poly_model.transform(poly_new_exp))
print("Polynomial Regression Prediction:\n",poly_pred)
print("MSE of Polynomial Regression:",mean_squared_error(Y,Y_pred,))
print("R2 Score of Polynomial regression:",r2_score(Y,Y_pred))
```
