**Linear Regression**

1. Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

2.Dataset

```python
data ={"Experience" : [1,2,3,4,5],
       "Salary" : [20000,30000,40000,50000,60000]}
df=pd.DataFrame(data)
```

3.Split data

```python
x=df[["Experience"]]
y=df["Salary"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

4.Train The Model

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

5.Prediction

```python
y_pred = model.predict(x_test)
print(y_pred)
```

6.Evaluation

```python
mse = mean_squared_error(y_test,y_pred)
#r2 = r2_score(y_test,y_pred)           #R2 is NaN becouse the test set contains only one sample
print("MSE:",mse)
#print("R2_score:",r2)                   #very small datasets, don't rely on r2, it requares at least two samples to compute veriance
```

*IMPORTANT ATTRIBUTES*

```python
print(model.coef_)               #Slope (m)
print(model.intercept_)          #intercept(c)
```

**Mini Task**

```python

data1 = {"Hours_Studied" : [1,2,3,4,5],
         "Marks" : [40,50,60,70,80]}
df=pd.DataFrame(data1)
#Data Split
X=df[["Hours_Studied"]]
Y=df["Marks"]
#Model
model1=LinearRegression()
model1.fit(X,Y)
#Prediction
predict_df = pd.DataFrame({"Hours_Studied":[6]})
predict=model1.predict(predict_df)
print("Predicted Marks for 6 hours:",predict[0])
#Slop
print(model1.coef_)
#Intercept
print(model1.intercept_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y,color="blue")
plt.plot(X,model1.predict(X),color="red")
plt.xlabel("Hours_Studied")
plt.ylabel("Marks")
plt.title("Simple LinearRegression")
plt.show()
```
