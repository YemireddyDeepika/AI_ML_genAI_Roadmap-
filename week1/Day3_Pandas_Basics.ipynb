```python
#Creating DataFrame
data = { 
         "Name":["A","B","c"],
         "Age":[22,25,23],
         "Marks":[85,90,78]
       }
df = pd.DataFrame(data)
print(df)

#Basic functions
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
#Practice Task
data = {"Students":["Subbu","Sree","Renu"],
        "Subjects":["Math","English","Social"],
        "Marks":[65,79,89]}
a = pd.DataFrame(data)
print(a)
#Printing 1st 2 rows and last row
print(a.head(2))
print(a.tail(1))
#Finding average and maximum marks
print(a["Marks"].mean())
print(a["Marks"].max())
```

```python
##MINI TASK##
b = {
    "Student": ["Ravi", "Anu", "Kiran", "Meena"],
    "Maths": [78, 85, 90, 66],
    "Science": [72, 88, 91, 70]
}

c = pd.DataFrame(b)
#Display the data
print(c)
#Finding average of maths and science
print("Average of math and science is:",c[["Maths","Science"]].mean())
#Topper in  maths
topper = c.loc[c["Maths"].idxmax(),["Student","Maths"]]
print("Topper in maths is:",topper)
```
