```python
import numpy as np
```

```python
#1D array
a = np.array([1,2,3,4,5])
print(a)
print("Shape:",a.shape)
```

```python
#2D array
b = np.array([[1,2,3],[4,5,6]])
print(b)
print("Shape:",b.shape)
```

```python
#Indexing
print(a[0])
print(b[1,2])
```

```python
#Slicing
print(a[1:4])
print(b[:,2])
```

```python
#Broadcasting
c = a+5
print(c)
print(b+5)
```
