```python
import pandas as pd
```

```python
data = {"Name" : ["Ravi","Anu","Kiran","Meena"],
        "Age" : [22,25,23,21],
        "Marks" : [78,85,90,66]}
df = pd.DataFrame(data)
print(df)
```

```python
#Column Selection
print(df["Name"])
print(df[["Name","Marks"]])
```

```python
#Row Selection
print(df.loc[0])
print(df.iloc[1:3,0])    #select row at integer position 1 and 2and the column position 0
print(df.iloc[1])
```

```python
#Conditional Filtering
print(df[df["Marks"]>80])
```

```python
#Sorting
print(df.sort_values(by="Marks",ascending=False))
```

```python
#Practice Tasks
print(df["Age"])              #Selecting only Age row
print(df.loc[0:1])            #select first 2 rows
print(df[df["Marks"]<70])     #Filter the students marks lessthan 70
print(df.sort_values(by="Age"))  #Sort by age
```

```python
#Mini Task
data1 = { "Employee":["A","B","C","D"],
          "Salary" : [30000,45000,25000,50000],
          "Experience" : [1,2,3,4]}
a = pd.DataFrame(data1)
print(a)
print(a[a["Salary"]>40000])                                 #employees with salary >40000
print(a.sort_values(by = "Experience" , ascending = False)) #sort by experience
print(a[["Employee","Salary"]])                             #Select employee and experience
```
