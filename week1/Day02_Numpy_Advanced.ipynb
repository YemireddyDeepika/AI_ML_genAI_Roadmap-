```python
import numpy as np
```

```python
#Create array
a = np.array([[1,2,3],[4,5,6]])
print(a)
```

```python
#Reshaped array
b= a.reshape(3,2)
print(b)
```

```python
#Flatten
print("Flatten:",a.flatten())
```

```python
#ravel 
c = a.ravel()
c[0] = 99
print(a)
```

```python
#Boolean masking
arr = np.array([10,3,7,2,15])
mask = arr>5                                  #mask can caonvert that condition true that elements is  othre wise false
print("Greater than 5:",arr[arr>5])
print("Greater than 5 :",mask)
```

```python
#Statistics
marks = np.array([[70,40,50],[79,39,50]])
print("Average marks:",marks.mean())
print("Row wise mean:",marks.mean(axis=1))
print("Column wise mean:",marks.mean(axis=0))
```

```python
##Pratice Task##
score = np.array([[3,8,6],[4,7,9],[2,18,6],[4,7,3]])
#reshape
print("Reshaped array:",score.reshape(3,4))
#Replace all values >5 with 0
score[score>5] = 0
print(score)
#Finding Max of row and Min of column
print("Max of row is:",score.max(axis=1))
print("Min of column is:",score.min(axis=0))
```
