```python
import pandas as pd
import numpy as np
```

```python
data = { "Name" : ["Ravi","Anu","Kiran","Meena"],
         "Age" : [32,np.nan,23,21],
         "Marks" : [78,85,np.nan,66]}
df = pd.DataFrame(data)
print(df)
print(df.info())
print(df.dropna())
#Check Missing Values
print(df.isnull())
print(df.isnull().sum())

#Fill Missing Values
df["Age"]=df["Age"].fillna(df["Age"].mean())
df["Marks"]=df["Marks"].fillna(df["Marks"].median())
print(df)
```
