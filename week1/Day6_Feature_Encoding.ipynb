```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

1.Label Encoding Example

```python
data = { "Level":["Low","Medium","High","Medium","Low"]}
df = pd.DataFrame(data)
print(df)
le=LabelEncoder()
df["Level_Encoder"]=le.fit_transform(df["Level"])
print(df)
```

2.One-Hot Encoder Example


```python
data1={"City":["Hybd","Bglr","Chennai","Hybd"]}
df1=pd.DataFrame(data1)
encoded_df1 = pd.get_dummies(df1,columns=["City"])                     #Create boolean columns by default when the input is string
encoded_df2 = pd.get_dummies(df1,columns=["City"],dtype=int)           #Converting data type int
encoded_df3 = pd.get_dummies(df1,columns=["City"],drop_first=True,dtype=int) #Avoid Dummies Variable Trap
print(encoded_df1)
print(encoded_df2)
print(encoded_df3)
```

Mini Task

```python
data2={"Gender":["Male","Female","Female","Male"],
       "Purchased":["Yes","No","Yes","No"]}
df4=pd.DataFrame(data2)
#Lable Encode Purchased
df4["Purchased"]=le.fit_transform(df4["Purchased"])

#One-Hot encoder Gender
df4=pd.get_dummies(df4,columns=["Gender"],drop_first=True,dtype=int)
print(df4)
```
