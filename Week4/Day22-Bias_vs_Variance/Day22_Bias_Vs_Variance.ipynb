**Mini Task**

Dataset

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

x,y = make_regression(n_samples=100,noise=20,random_state=42)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

Case 1: High Bias Model(underfitting)

```python
model_low = DecisionTreeRegressor(max_depth=1)
model_low.fit(x_train,y_train)

train_error = mean_squared_error(y_train,model_low.predict(x_train))
test_error = mean_squared_error(y_test,model_low.predict(x_test))

print("High Bias Model")
print("Train Error:",train_error)
print("Test error:",test_error)
```

Case 2:High Variance Model

```python
model_high = DecisionTreeRegressor(max_depth=20)
model_high.fit(x_train,y_train)

train_error = mean_squared_error(y_train,model_high.predict(x_train))
test_error = mean_squared_error(y_test,model_high.predict(x_test))

print("High Variance model")
print("Train Error:",train_error)
print("Test Error:",test_error)
```
