```python
import pandas as pd
```

MinMax Scaler Example

```python
from sklearn.preprocessing import MinMaxScaler
data={
    "Age":[20,30,40,50],
    "Salary":[20000,30000,40000,50000]}
df=pd.DataFrame(data)
scale=MinMaxScaler()
scaled = scale.fit_transform(df)
scaled_df1=pd.DataFrame(scaled,columns=df.columns)
print(scaled_df1)
```

Standerd Scaler Example


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data=scaler.fit_transform(df)
scaled_df=pd.DataFrame(scaled_data,columns=df.columns)
print(scaled_df)
```

Mini Task

```python
data1={
    "Class":[2,5,7,9],
    "Feez":[25000,30000,45000,80000]}
df1=pd.DataFrame(data1)
#Standard Scaling
stan_scale=StandardScaler()
standard=stan_scale.fit_transform(df1)
stan_df1=pd.DataFrame(standard,columns=df1.columns)
print("\nStandard Scaled Data")
print(stan_df1)

#MinMax Scaling
minmax_scale=MinMaxScaler()
minmax=minmax_scale.fit_transform(df1)
minmax_df1=pd.DataFrame(minmax,columns=df1.columns)
print("\nMinMax Scaled Data")
print(minmax_df1)
```
