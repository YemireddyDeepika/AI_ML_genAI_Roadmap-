**Activation Functions**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([-2, -1, 0, 1, 2])

#Sigmoid
sigmoid = 1 / (1 + np.exp(-x))

#ReLU
relu = np.maximum(0, x)

print("Input:",x)
print("Sigmoid:", sigmoid)
print("ReLU:", relu)
```

```python
#Plot the activation function
plt.plot(x, sigmoid, marker='o', label="Sigmoid")
plt.plot(x, relu, marker='o', label="ReLU")

plt.title("Sigmoid Vs ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```
