**Artificial Neural Network(ANN)**

1.Import Libraries

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
```

2.XOR Dataset

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])
```

3.Model

```python
model = MLPClassifier(hidden_layer_sizes=(4,4),
                      activation="tanh",
                      solver="lbfgs",
                      max_iter=1000,
                      random_state=42)
model.fit(x,y)
```

4.Prediction

```python
y_pred = model.predict(x)

acc = accuracy_score(y,y_pred)
print("Accuracy:",acc)
```

```python
print(model.predict([[0,0],[0,1],[1,0],[1,1]]))
```

5.Visualization

```python
#Create mesh grid
x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300))

#Pridict on every grid point
z = model.predict(np.c_[xx.ravel(),yy.ravel()])
z = z.reshape(xx.shape)

#plot decision boundary
plt.contour(xx, yy, z, alpha=0.3)

#plot XOR data points
plt.scatter(x[:,0], x[:,1], c=y, s=120, edgecolors="black")

#Labels & Titles
plt.title("MLPClassifier decision Boundary(XOR Problem)")
plt.xlabel("Input 1")
plt.ylabel("Input 2")

plt.show()
```
